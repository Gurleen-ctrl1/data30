```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

primes = [str(i) for i in range(1, 101) if is_prime(i)]

with open("primes.txt", "w") as f:
    f.write("\n".join(primes))
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def average(self):
        return sum(self.marks) / len(self.marks)

s1 = Student("Aman", [85, 90, 80])
print(f"{s1.name}'s average: {s1.average()}")
```

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w
    def area(self):
        return self.l * self.w

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return math.pi * self.r ** 2

r = Rectangle(5, 3)
c = Circle(4)
print("Rectangle area:", r.area())
print("Circle area:", round(c.area(), 2))
```
